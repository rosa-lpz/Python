# **Import libraries**


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# **Datos**

## Datos de vuelos

```python
# initialize list of lists
data = [['1949', 'Jan', 112], ['1949', 'Feb', 118], ['1949', 'Mar', 132], ['1949', 'Apr', 129], ['1949', 'May', 108],
        ['1950', 'Jan', 110], ['1950', 'Feb', 121], ['1950', 'Mar', 112], ['1950', 'Apr', 138], ['1950', 'May', 113],
        ['1951', 'Jan', 129], ['1951', 'Feb', 134], ['1951', 'Mar', 118], ['1951', 'Apr', 111], ['1951', 'May', 125],
        ['1952', 'Jan', 111], ['1952', 'Feb', 125], ['1952', 'Mar', 133], ['1952', 'Apr', 121], ['1952', 'May', 115],
        ['1953', 'Jan', 125], ['1953', 'Feb', 133], ['1953', 'Mar', 111], ['1953', 'Apr', 132], ['1953', 'May', 121],
        ['1954', 'Jan', 129], ['1954', 'Feb', 112], ['1954', 'Mar', 112], ['1954', 'Apr', 129], ['1954', 'May', 118],
        ['1955', 'Jan', 133], ['1955', 'Feb', 118], ['1955', 'Mar', 128], ['1955', 'Apr', 122], ['1955', 'May', 132]]

# Create the pandas DataFrame
flights = pd.DataFrame(data, columns=['year', 'month','passengers'])

# print dataframe.
flights.head()
```

# **Line plot (Gr√°fico lineal)**

```python
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

### Adjust size of the plot

```python
sns.set(rc={"figure.figsize":(10, 8)}) # To adjust size
```

```python

may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

# **Bar plot (Gr√°fico de barras)**

```python
sns.set_theme(style="whitegrid")
sns.barplot(x="month",y="passengers",hue="year",data=flights)
```

# **Histograma**

```python
sns.histplot(data=flights, x="passengers")
```

```python
sns.histplot(data=flights, x="passengers", bins=5)
```

```python
sns.histplot(data=flights, x="passengers", hue='year')
```

# **Displot (Distribution plot)**

```python
sns.displot(data=flights, x="passengers", hue='year', multiple="stack")
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(data=flights, x="passengers", hue='year')
```

# **Heatmap**

References

```python
sns.set()
flights2 = flights.pivot("month", "year", "passengers")
sns.set(rc={"figure.figsize":(10, 8)})
ax = sns.heatmap(flights2)
plt.title("Heatmap Flight Data")
plt.show()
```

# Pie Diagram

## Example 1

```python
# Group by 'year' and sum the passengers for each year
yearly_passengers = flights.groupby('year')['passengers'].sum()
yearly_passengers
```

```python
# Apply Seaborn's default style
sns.set_theme()

# Create a pie chart using Matplotlib's pie function
plt.figure(figsize=(8, 8))
colors = sns.color_palette("Set2", len(yearly_passengers))  # Use Seaborn's color palette

# Plot the pie chart
plt.pie(yearly_passengers, labels=yearly_passengers.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'})
plt.title('Distribution of Passengers by Year')
plt.ylabel('')  # Remove the y-label
plt.show()
```

# Scatter Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Simple scatter plot

```python
x = [2016,2017,2018,2019,2020,2021]
y = [45,46,47,48,50,51]

plt.scatter(x=x, y=y)
plt.xlabel('Years')
plt.ylabel('Population (M)')
plt.title ('Years vs Population')

plt.show()
```

## Scatter plot - labels, marker & color personalization

```python
plt.scatter(x=x, y=y, marker='*', color='r')
plt.xlabel('Years')
plt.ylabel('Population (M)')
plt.title ('Years vs Population')

plt.show()
```

## Scatter plot - Size customization

```python
plt.figure(figsize=(10,6))
plt.scatter(x=x, y=y, marker='*', color='r')
plt.xlabel('Years')
plt.ylabel('Population (M)')
plt.title ('Years vs Population')

plt.show()
```

# Bar Charts

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Simple Bar chart

```python
# Example 1
x2 = ['Argentina','Colombia','Peru']
y2 = [40,50,33]

plt.bar(x2,y2)
plt.show()
```

```python
# Example 2
movies = ["Annie Hall", "Ben Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars)

# Etiquetas
plt.xlabel("Película") # Eje X
plt.ylabel(" # Premios de la academia") # Eje Y
plt.title("Mis películas favoritas") # Título del gráfico


plt.show()
```

# Bar chart - color customization

```python
plt.figure(figsize=(10,6))

plt.bar(x2,y2, color='g')
plt.xlabel('Country')
plt.ylabel('Population (Millons)')
plt.title ('Country vs Population')

plt.show()
```

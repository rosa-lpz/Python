# Line Charts

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Simple Line Chart

```python
x = [2016,2017,2018,2019,2020,2021]
y = [45,46,47,48,50,51]

plt.plot(x,y)
plt.show()
```

## Line Chart - labels, marker, linestyle, color customization

```python
plt.plot(x,y, marker='o', linestyle='--', color='r')
plt.xlabel('Years')
plt.ylabel('Population (M)')
plt.title ('Years vs Population')

plt.show()
```

## Line chart - Size customization

```python
plt.figure(figsize=(10,6))


#x = [2016,2017,2018,2019,2020,2021]
#y = [45,46,47,48,50,51]

plt.plot(x,y, marker='o', linestyle='--', color='r')
plt.xlabel('Years')
plt.ylabel('Population (M)')
plt.title ('Years vs Population')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig.tight_layout()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Time series DF creation

    * np.random.randn(100) generates 100 random numbers from a standard normal distribution.
    * .cumsum() calculates the cumulative sum to simulate a trend over time (more realistic for time series).

```python
# Create a time series index
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum()
```

```python
dates
```

```python
values
```

```python
# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value': values})
df.set_index('Date', inplace=True)
df
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Value', color='blue')
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

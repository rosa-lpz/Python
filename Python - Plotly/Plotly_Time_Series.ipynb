```python

import plotly.express as px
import pandas as pd

# Sample time series dataset
data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Value': [10, 12, 9, 15, 11]
}
df = pd.DataFrame(data)

# Time series plot using Plotly Express
fig = px.line(df, x='Date', y='Value', title='Sample Time Series Data')

# Display the plot
fig.show()
```

```python

import plotly.express as px
import pandas as pd
import numpy as np

# Sample time series dataset with 500 rows
start_date = pd.to_datetime('2022-01-01')
end_date = pd.to_datetime('2025-01-01')
dates = pd.to_datetime(np.random.uniform(start_date.value, end_date.value, 500), unit='ns')
dates = dates.sort_values() # Ensure dates are in order for the time series plot

# Random values
values = np.random.rand(500) * 100

data = {
    'Date': dates,
    'Value': values
}
df = pd.DataFrame(data)

# Time series plot using Plotly Express
fig = px.line(df, x='Date', y='Value', title='Sample Time Series Data (500 rows)')

# Display the plot
fig.show()
```
